```python
from PIL import Image
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 打开图片
image_path = r"D:\1.png"
image = Image.open(image_path)

# 获取图片的像素数据
pixels = image.load()

# 图片尺寸
width, height = image.size

# 创建3D坐标系
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 计算z值之和
z_sum = 0

# 遍历每个像素点
for x in range(width):
    for y in range(height):
        # 获取像素的RGB值
        r, g, b = pixels[x, y]

        # 判断是否为红色（假设阈值为100，你可以根据实际情况进行调整）
        if r > 100 and g < 100 and b < 100:
            z = 1
            z_sum += z
            ax.scatter(x, y, z, c='red')  # 在3D坐标系中添加红色点
        else:
            z = 0
            ax.scatter(x, y, z, c='blue')  # 在3D坐标系中添加蓝色点

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 显示图形
plt.show()

# 打印z值之和
print("Sum of z values:", z_sum)
```

```python
from PIL import Image
import numpy as np
from sklearn.cluster import KMeans

# 打开图像并加载像素
image_path1 = r"D:\Adelaide\Trimester\2023 Trimester 3\MATHS 7097B - Data Science Res Proj Pt B\ProjectB作战报告\图像标点图\day3col5row2.jpg"
image = Image.open(image_path1)
pixels = image.load()
width, height = image.size

# 初始化一个列表来存储红色标记点的坐标
red_points = []

# 遍历图像的每个像素
for x in range(width):
    for y in range(height):
        color = pixels[x, y]
        # 检查像素是否属于红色范围
        if color[0] >= 240 and color[0] <= 255 and color[1] <= 5 and color[2] <= 5:
            red_points.append((x, y))

# 将红色标记点的x值提取出来
red_points_x = [point[0] for point in red_points]

# 使用KMeans算法将x值分成6个簇
kmeans = KMeans(n_clusters=6)
kmeans.fit(np.array(red_points_x).reshape(-1, 1))

# 获取每个点所属的簇和x值
cluster_labels = kmeans.labels_
points_with_x = [(point, x) for point, x in zip(red_points, red_points_x)]

# 将点按簇分组，并按照x值排序
clusters = {}
for i, label in enumerate(cluster_labels):
    if label not in clusters:
        clusters[label] = []
    clusters[label].append(points_with_x[i])

# 对每个簇中的点按x值排序
sorted_clusters = sorted(clusters.items(), key=lambda item: min(item[1], key=lambda x: x[1])[1])

# 打印每个簇的所有点的坐标，同时按照x值排序的顺序命名簇
for cluster_index, (label, points) in enumerate(sorted_clusters):
    sorted_points = sorted(points, key=lambda x: x[1])
    print(f"Cluster {cluster_index + 1} Points (Sorted by x value):")
    for point, _ in sorted_points:
        print(point)
```

```python
import csv

# 定义一个函数来计算每个簇的矩形范围和统计点数
def calculate_cluster_rectangles(clusters, image_height):
    results = []
    # 创建一个集合来跟踪已经计数的点
    counted_points = set()

    for cluster_name, cluster_points in clusters:
        x_values = [point[1] for point in cluster_points]
        min_x = min(x_values) - 1
        max_x = max(x_values) + 1
        x_range = f"{min_x}-{max_x}"
        y_range = f"0-{image_height}"  # 使用整个图像的y范围

        # 将y范围分成四份
        y_ranges = [(i * (image_height // 4), (i + 1) * (image_height // 4)) for i in range(4)]

        # 统计每份的标记点个数
        counts = [0] * 4
        for point in cluster_points:
            for i, (y_min, y_max) in enumerate(y_ranges):
                if y_min <= point[0][1] <= y_max and point[0] not in counted_points:
                    counts[i] += 1
                    counted_points.add(point[0])  # 标记已经计数的点

        # 将结果添加到results列表
        for i, (y_min, y_max) in enumerate(y_ranges):
            results.append([f"Cluster {cluster_name}", f"Partition {i+1}", counts[i]])

    return results

# 计算簇的矩形范围和点数
results = calculate_cluster_rectangles(sorted_clusters, height)

# 保存结果到CSV文件
csv_file_path = "cluster_results.csv"
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Cluster Name", "Partition Name", "Count"])
    writer.writerows(results)

# 打开CSV文件
import os
os.system(f'start excel.exe {csv_file_path}')
```

```python
import csv
import subprocess
import shutil
from PIL import Image
import numpy as np
from sklearn.cluster import KMeans
import os

# 指定图像文件夹路径
image_folder_path = r"D:\Adelaide\Trimester\2023 Trimester 3\MATHS 7097B - Data Science Res Proj Pt B\ProjectB作战报告\图像标点图"

# 指定CSV文件保存的目录
csv_destination_path = r"D:\Adelaide\Trimester\2023 Trimester 3\MATHS 7097B - Data Science Res Proj Pt B\ProjectB作战报告\新数据"

# 获取文件夹中所有.jpg文件的路径
image_paths = [os.path.join(image_folder_path, filename) for filename in os.listdir(image_folder_path) if filename.endswith(".jpg")]

def calculate_cluster_rectangles(clusters, image_height):
    results = []
    # 创建一个集合来跟踪已经计数的点
    counted_points = set()

    for cluster_name, cluster_points in clusters:
        x_values = [point[1] for point in cluster_points]
        min_x = min(x_values) - 1
        max_x = max(x_values) + 1
        x_range = f"{min_x}-{max_x}"
        y_range = f"0-{image_height}"  # 使用整个图像的y范围

        # 将y范围分成四份
        y_ranges = [(i * (image_height // 4), (i + 1) * (image_height // 4)) for i in range(4)]

        # 统计每份的标记点个数
        counts = [0] * 4
        for point in cluster_points:
            for i, (y_min, y_max) in enumerate(y_ranges):
                if y_min <= point[0][1] <= y_max and point[0] not in counted_points:
                    counts[i] += 1
                    counted_points.add(point[0])  # 标记已经计数的点

        # 将结果添加到results列表
        for i, (y_min, y_max) in enumerate(y_ranges):
            results.append([f"Cluster {cluster_name}", f"Partition {i+1}", counts[i]])

    return results

# 遍历每个图像文件
for image_path in image_paths:
    # 打开图像并加载像素
    image = Image.open(image_path)
    pixels = image.load()
    width, height = image.size

    # 初始化一个列表来存储红色标记点的坐标
    red_points = []

    # 遍历图像的每个像素
    for x in range(width):
        for y in range(height):
            color = pixels[x, y]
            # 检查像素是否属于红色范围
            if color[0] >= 240 and color[0] <= 255 and color[1] <= 5 and color[2] <= 5:
                red_points.append((x, y))

    # 将红色标记点的x值提取出来
    red_points_x = [point[0] for point in red_points]

    # 使用KMeans算法将x值分成6个簇
    kmeans = KMeans(n_clusters=6)
    kmeans.fit(np.array(red_points_x).reshape(-1, 1))

    # 获取每个点所属的簇和x值
    cluster_labels = kmeans.labels_
    points_with_x = [(point, x) for point, x in zip(red_points, red_points_x)]

    # 将点按簇分组，并按照x值排序
    clusters = {}
    for i, label in enumerate(cluster_labels):
        if label not in clusters:
            clusters[label] = []
        clusters[label].append(points_with_x[i])

    # 对每个簇中的点按x值排序
    sorted_clusters = sorted(clusters.items(), key=lambda item: min(item[1], key=lambda x: x[1])[1])
    # 计算簇的矩形范围和点数
    results = calculate_cluster_rectangles(sorted_clusters, height)

    # 构建CSV文件名
    csv_filename = os.path.splitext(os.path.basename(image_path))[0] + ".csv"
    
    # 创建CSV文件的完整路径
    csv_full_path = os.path.join(csv_destination_path, csv_filename)
    with open(csv_full_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["Cluster Name", "Partition Name", "Count"])
        writer.writerows(results)
```
